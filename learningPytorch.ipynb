Build and Train a Simple Feedforward Network on MNIST

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt

# =============================
# 1. Dataset
# =============================
transform = transforms.ToTensor()
train_data = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_data  = datasets.MNIST(root='./data', train=False, transform=transform)

train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader  = DataLoader(test_data, batch_size=64, shuffle=False)

# =============================
# 2. Model Definition
# =============================
class FeedforwardNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(28*28, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = x.view(-1, 28*28)
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = FeedforwardNN().to(device)

# =============================
# 3. Loss & Optimizer
# =============================
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# =============================
# 4. Training Loop
# =============================
epochs = 5
train_losses = []

for epoch in range(epochs):
    running_loss = 0
    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    avg_loss = running_loss/len(train_loader)
    train_losses.append(avg_loss)
    print(f"Epoch [{epoch+1}/{epochs}] Loss: {avg_loss:.4f}")

plt.plot(train_losses, label='Training Loss')
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.title("MNIST Training Loss Curve")
plt.legend(); plt.show()
```

Compare Optimizers (SGD vs Adam vs RMSProp)

```python
optimizers = {
    "SGD": optim.SGD(model.parameters(), lr=0.01),
    "Adam": optim.Adam(model.parameters(), lr=0.001),
    "RMSProp": optim.RMSprop(model.parameters(), lr=0.001)
}

results = {}

for name, opt in optimizers.items():
    model = FeedforwardNN().to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = opt
    losses = []

    for epoch in range(10):  # quick test
        running_loss = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        losses.append(running_loss/len(train_loader))
        print(f"{name} - Epoch [{epoch+1}/{epochs}] Loss: {running_loss/len(train_loader):.4f}")
    results[name] = losses


for name, losses in results.items():
    plt.plot(losses, label=name)
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.title("Optimizer Comparison")
plt.legend(); plt.show()
```

Activation Function Experiment

```python
activations = {
    "ReLU": nn.ReLU(),
    "LeakyReLU": nn.LeakyReLU(0.01),
    "Sigmoid": nn.Sigmoid(),
    "Tanh": nn.Tanh()
}

results = {}

for name, act in activations.items():
    class CustomNN(nn.Module):
        def __init__(self):
            super().__init__()
            self.fc1 = nn.Linear(28*28, 128)
            self.fc2 = nn.Linear(128, 64)
            self.fc3 = nn.Linear(64, 10)
            self.act = act
        def forward(self, x):
            x = x.view(-1, 28*28)
            x = self.act(self.fc1(x))
            x = self.act(self.fc2(x))
            return self.fc3(x)

    model = CustomNN().to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    losses = []
    for epoch in range(10):
        total_loss = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            out = model(images)
            loss = criterion(out, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        losses.append(total_loss/len(train_loader))
        print(f"{name} - Epoch [{epoch+1}/{10}] Loss: {total_loss/len(train_loader):.4f}")
    results[name] = losses

for name, losses in results.items():
    plt.plot(losses, label=name)
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.title("Activation Function Comparison")
plt.legend(); plt.show()
```

Learning Rate Scheduler (Dynamic LR)

```python
model = FeedforwardNN().to(device)
optimizer = optim.Adam(model.parameters(), lr=0.1)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=2, gamma=0.5)
criterion = nn.CrossEntropyLoss()

lr_values = []
losses = []

for epoch in range(10):
    total_loss = 0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        out = model(images)
        loss = criterion(out, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    scheduler.step()
    lr_values.append(optimizer.param_groups[0]['lr'])
    losses.append(total_loss/len(train_loader))
    print(f"Epoch [{epoch+1}/{6}] Loss: {total_loss/len(train_loader):.4f} - LR: {optimizer.param_groups[0]['lr']}")

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(losses, 'g-', label='Loss')
ax2.plot(lr_values, 'b--', label='Learning Rate')
ax1.set_xlabel("Epochs"); ax1.set_ylabel("Loss (green)"); ax2.set_ylabel("LR (blue)")
plt.title("Learning Rate Scheduler Effect"); plt.show()
```

Loss Function Comparison (CrossEntropy vs MSE)

```python
def train_with_loss(loss_fn, epochs=10):
    model = FeedforwardNN().to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    losses = []
    for epoch in range(epochs):
        total_loss = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            if isinstance(loss_fn, nn.MSELoss):
                labels_onehot = torch.nn.functional.one_hot(labels, num_classes=10).float()
                loss = loss_fn(torch.softmax(outputs, dim=1), labels_onehot)
            else:
                loss = loss_fn(outputs, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        losses.append(total_loss/len(train_loader))
        print(f"Epoch [{epoch+1}/{epochs}] Loss: {total_loss/len(train_loader):.4f}")
    return losses

mse_loss = train_with_loss(nn.MSELoss())
ce_loss  = train_with_loss(nn.CrossEntropyLoss())

plt.plot(mse_loss, label="MSELoss")
plt.plot(ce_loss, label="CrossEntropyLoss")
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.title("Loss Function Comparison")
plt.legend(); plt.show()
```

```python
# Lab 9: Build and Train Your First CNN on CIFAR-10
# --------------------------------------------------
# Run this section on Colab (GPU recommended)
# Make sure "Runtime > Change runtime type > GPU" is selected

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# Data transformation: normalize images between [-1, 1]
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5),
                         (0.5, 0.5, 0.5))
])

# Load training and test data
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64,
                                          shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64,
                                         shuffle=False, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
class SimpleCNN(nn.Module):
    def __init__(self, kernel_size=3, stride=1, pool_size=2):
        super(SimpleCNN, self).__init__()
        # Convolution layers
        self.conv1 = nn.Conv2d(3, 16, kernel_size=kernel_size, stride=stride, padding=1)
        self.pool = nn.MaxPool2d(pool_size, pool_size)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=kernel_size, stride=stride, padding=1)

        # Fully connected layers
        self.fc1 = nn.Linear(32 * 8 * 8, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))   # Conv1 + ReLU + Pool
        x = self.pool(F.relu(self.conv2(x)))   # Conv2 + ReLU + Pool
        x = x.view(-1, 32 * 8 * 8)             # Flatten
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Instantiate model
model = SimpleCNN(kernel_size=3, stride=1, pool_size=2).to(device)


# Example experiments:
# (1) Larger stride → fewer features, faster training
model = SimpleCNN(kernel_size=3, stride=2, pool_size=2).to(device)

# (2) Smaller kernel (1x1) → focus on color/intensity features
model = SimpleCNN(kernel_size=1, stride=1, pool_size=2).to(device)

# (3) Larger kernel (5x5) → detects more global features
model = SimpleCNN(kernel_size=5, stride=1, pool_size=2).to(device)
print(model)

```

```python
class VisualCNN(nn.Module):
    def __init__(self, kernel_size=3, stride=1, pool_size=2):
        super(VisualCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size, stride, padding=1)
        self.pool = nn.MaxPool2d(pool_size, pool_size)
        self.conv2 = nn.Conv2d(16, 32, kernel_size, stride, padding=1)

        # 👇 Compute final feature map size dynamically using a dummy input
        with torch.no_grad():
            x = torch.zeros(1, 3, 32, 32)
            x = self.pool(F.relu(self.conv1(x)))
            x = self.pool(F.relu(self.conv2(x)))
            flatten_size = x.numel()

        self.fc1 = nn.Linear(flatten_size, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        print(f"Input: {x.shape}")
        x = F.relu(self.conv1(x))
        print(f"After Conv1: {x.shape}")
        x = self.pool(x)
        print(f"After Pool1: {x.shape}")
        x = F.relu(self.conv2(x))
        print(f"After Conv2: {x.shape}")
        x = self.pool(x)
        print(f"After Pool2: {x.shape}")
        x = torch.flatten(x, 1)
        print(f"After Flatten: {x.shape}")
        x = F.relu(self.fc1(x))
        print(f"After FC1: {x.shape}")
        x = self.fc2(x)
        print(f"Output: {x.shape}")
        return x
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 50
train_losses = []

for epoch in range(num_epochs):
    running_loss = 0.0
    for i, (images, labels) in enumerate(trainloader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(trainloader)
    train_losses.append(epoch_loss)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}")

print("Training complete.")
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(32 * 8 * 8, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 32 * 8 * 8)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = SimpleCNN().to(device)
print(model)
```

DEEPER CNN

```python
class DeepVisualCNN(nn.Module):
    def __init__(self):
        super(DeepVisualCNN, self).__init__()

        # Convolutional feature extractor
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)     # Output: 16 x 32 x 32
        self.bn1 = nn.BatchNorm2d(16)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)    # Output: 32 x 32 x 32
        self.bn2 = nn.BatchNorm2d(32)
        self.pool1 = nn.MaxPool2d(2, 2)                 # Output: 32 x 16 x 16

        self.conv3 = nn.Conv2d(32, 64, 3, padding=1)    # Output: 64 x 16 x 16
        self.bn3 = nn.BatchNorm2d(64)
        self.pool2 = nn.MaxPool2d(2, 2)                 # Output: 64 x 8 x 8

        self.conv4 = nn.Conv2d(64, 128, 3, padding=1)   # Output: 128 x 8 x 8
        self.bn4 = nn.BatchNorm2d(128)

        self.conv5 = nn.Conv2d(128, 256, 3, padding=1)  # Output: 256 x 8 x 8
        self.bn5 = nn.BatchNorm2d(256)
        self.pool3 = nn.MaxPool2d(2, 2)                 # Output: 256 x 4 x 4

        # Fully connected layers
        self.fc1 = nn.Linear(256 * 4 * 4, 512)
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(512, 10)

    def forward(self, x):
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = self.pool1(x)

        x = F.relu(self.bn3(self.conv3(x)))
        x = self.pool2(x)

        x = F.relu(self.bn4(self.conv4(x)))
        x = F.relu(self.bn5(self.conv5(x)))
        x = self.pool3(x)

        x = x.view(-1, 256 * 4 * 4)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

model = DeepVisualCNN().to(device)
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=0.001)  # Only train classifier
```

Training Loop **(Feature Extraction Phase)**

```python
def train_model(model, loader, criterion, optimizer, epochs=5):
    model.train()
    for epoch in range(epochs):
        running_loss = 0.0
        for images, labels in tqdm(loader, desc=f"Epoch {epoch+1}/{epochs}"):
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(loader):.4f}")
    print("Training complete.")

train_model(model, trainloader, criterion, optimizer, epochs=5)
```

```python
def evaluate_model(model, loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    acc = 100 * correct / total
    print(f"Accuracy: {acc:.2f}%")
    return acc

evaluate_model(model, testloader)
```

Fine-Tuning **(Unfreeze and Train Deeper Layers)**

```python
# Unfreeze last few layers for fine-tuning
for name, param in model.named_parameters():
    if "layer4" in name or "fc" in name:
        param.requires_grad = True

# Lower learning rate for fine-tuning
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-4)

train_model(model, trainloader, criterion, optimizer, epochs=5)
evaluate_model(model, testloader)
```

```python
from sklearn.metrics import accuracy_score

def evaluate(model, loader):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())
    acc = accuracy_score(y_true, y_pred)
    print(f"Accuracy: {acc * 100:.2f}%")
    return y_true, y_pred
```

Project the last-layer embeddings (e.g., before the final classifier) into 2D — see how **classes cluster!**

```python
from sklearn.manifold import TSNE
import numpy as np

def visualize_tsne(model, loader, n_samples=1000):
    model.eval()
    features, labels = [], []
    with torch.no_grad():
        for images, labs in loader:
            images, labs = images.to(device), labs.to(device)
            feats = model.avgpool(model.layer4(model.layer3(model.layer2(model.layer1(model.conv1(images))))))
            feats = torch.flatten(feats, 1)
            features.append(feats.cpu().numpy())
            labels.extend(labs.cpu().numpy())
            if len(labels) >= n_samples:
                break

    X = np.concatenate(features)[:n_samples]
    y = np.array(labels)[:n_samples]
    tsne = TSNE(n_components=2, random_state=42)
    X_embedded = tsne.fit_transform(X)

    plt.figure(figsize=(10,8))
    scatter = plt.scatter(X_embedded[:,0], X_embedded[:,1], c=y, cmap='tab10', s=10)
    plt.legend(handles=scatter.legend_elements()[0], labels=classes, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.title("t-SNE of ResNet Feature Embeddings")
    plt.show()

visualize_tsne(model, testloader)
```

```python
# ===============================
# LAB 12: Regularization (Dropout + L2)
# ===============================

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# -------------------------------
# 1. Data Preparation
# -------------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

trainloader = DataLoader(trainset, batch_size=64, shuffle=True)
testloader = DataLoader(testset, batch_size=64, shuffle=False)

# -------------------------------
# 2. Load Pretrained Model (ResNet18)
# -------------------------------
model = torchvision.models.resnet18(weights='IMAGENET1K_V1')

# Freeze feature extractor (for feature extraction)
for param in model.parameters():
    param.requires_grad = False

# Replace the classifier with dropout + FC layers
num_features = model.fc.in_features
model.fc = nn.Sequential(
    nn.Linear(num_features, 256),
    nn.ReLU(),
    nn.Dropout(p=0.5),         # Dropout Regularization
    nn.Linear(256, 10)
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# -------------------------------
# 3. Define Loss and Optimizer (with L2 Regularization)
# -------------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)  # L2 Regularization

# -------------------------------
# 4. Training Loop
# -------------------------------
num_epochs = 5
train_losses, test_losses = [], []

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    train_losses.append(running_loss / len(trainloader))

    # Evaluate on test set
    model.eval()
    test_loss = 0
    correct, total = 0, 0
    with torch.no_grad():
        for images, labels in testloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            test_loss += loss.item()

            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = 100 * correct / total
    test_losses.append(test_loss / len(testloader))

    print(f"Epoch [{epoch+1}/{num_epochs}] - Train Loss: {train_losses[-1]:.4f} | Test Loss: {test_losses[-1]:.4f} | Test Acc: {acc:.2f}%")

# -------------------------------
# 5. Visualization
# -------------------------------
plt.figure(figsize=(8,5))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.title("Training vs Test Loss with Regularization")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
